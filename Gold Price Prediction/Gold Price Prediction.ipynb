## Exploring the Data
Run the code cell below to load necessary Python libraries and load the kaggle data.

```python
# Import libraries necessary for this project
import numpy as np
import pandas as pd
from time import time
from IPython.display import display # Allows the use of display() for DataFrames
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')

# Pretty display for notebooks
%matplotlib inline

# Load the Kaggle dataset
myData = pd.read_csv("FINAL_USO.csv")

# Success - Display the first record
display(myData.head(n=10))
myData.info()
myData.isnull().sum()
```

```python
myData.describe()
```

```python
# We will start out by selecting features gold ETF features
gold_features = ['Open','High', 'Low', 'Volume']
Features = myData[gold_features]
display(Features.head())

target = myData['Adj Close'] 
```

# Test Linear Regression Model for small data

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
MyGoldModel = LinearRegression()

#Fit Model
MyGoldModel.fit(Features, Target)

print("Making predicitons for the first 5 entries\n")
print(Features.head())
print("\nThe predictions are:\n")
print(MyGoldModel.predict(Features.head()))
print("\nThe actual values are:\n")
print(target.head())

predicted_adj_close = MyGoldModel.predict(Features.head())
print(mean_absolute_error(target.head(),predicted_adj_close))
```

As we can predict the Target easly by using inly five main features so we will expand this model and train it completely

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler


# Partition data into training and validation groups
X_train, X_test, y_train, y_test = train_test_split(Features, target, random_state = 0)

# Show the results of the split
print("Training set has {} samples.".format(X_train.shape[0]))
print("Testing set has {} samples.".format(X_test.shape[0]))


# Define a new model for training set
gold_model = LinearRegression()
# Fit model
gold_model.fit(X_train, y_train)

#get predicted prices on validation data
val_predictions = gold_model.predict(X_test)
print(mean_absolute_error(y_test,val_predictions))


Final_Model = make_pipeline(StandardScaler(), LinearRegression()).fit(X_train, y_train)

Target = Final_Model.predict(X_test)
print(Target[:10])

RMS = np.sqrt(mean_squared_error(y_test, Target))

yhat = pd.Series(Target, index = y_test.index)

Target.plot.line(label = 'Predicted Value')
y_test.plot.line(label = 'Real Value')
plt.legend()
plt.show()
```
